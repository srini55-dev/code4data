# filtering and ordering

```python
df[df['Rank'] <= 10]
```

```python
# sql: in condition 
specific_countries = ['Bangladesh', 'Pakistan' ]
df[df['Country'].isin(specific_countries)]
```

```python
# sql: like condition ( country like ('%United%'))
df[df['Country'].str.contains('United')]
```

```python
df1 = df.set_index('Country')
df1.head(2)
```

```python
# df1[['Continent','CCA3']]
# another alternative way is to use filter. filter() is used to filter based on index labels (row names) or column names, not on values.\
df1.filter(items=['Continent','CCA3'], axis = 1)
```

```python
# df1.loc[['Afghanistan','Albania']]
# another alternative to loc for selecting label based index rows is we can use filter with axis = 0
df1.filter(items=['Afghanistan','Albania'], axis = 0)
```

```python
# we can also use like similar to like opeartor in sql. this will select all rows that matches 'United' index 
df1.filter(like = 'United', axis = 0)
```

```python
# sorting
df1.sort_values(by=['Continent','Country'])
```
