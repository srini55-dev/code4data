# group by and aggregating

```python
import pandas as pd

# Sample DataFrame
data = {'Category': ['A', 'B', 'A', 'C', 'B', 'A'],
        'Value1': [10, 15, 12, 8, 20, 11],
        'Value2': [1, 2, 1, 3, 2, 1]}
df = pd.DataFrame(data)

# Group by 'Category' and calculate the sum of 'Value1'
result1 = df.groupby('Category')['Value1'].agg('sum')
print("Sum of Value1 per Category:\n", result1)

# Group by 'Category' and calculate multiple aggregations on 'Value1'
result2 = df.groupby('Category')['Value1'].agg(['sum', 'mean', 'count'])
print("\nMultiple aggregations on Value1 per Category:\n", result2)

# Group by 'Category' and apply different aggregations to different columns
result3 = df.groupby('Category').agg({'Value1': 'sum', 'Value2': 'mean'})
print("\nDifferent aggregations on different columns:\n", result3)

# Group by 'Category' and apply multiple aggregations to different columns
result4 = df.groupby('Category').agg({'Value1': ['sum', 'max'], 'Value2': 'mean'})
print("\nMultiple aggregations on different columns:\n", result4)

# Using named aggregation for custom column names in the output
result5 = df.groupby('Category').agg(
    total_value1=('Value1', 'sum'),
    avg_value2=('Value2', 'mean')
)
print("\nNamed aggregation:\n", result5)
```

# Merge, Join, and Concatenate

```python
import pandas as pd

# Define the data as a list of dictionaries
data1 = [
    {"FellowshipID": 1001, "FirstName": "Frodo",   "Skills": "Hiding"},
    {"FellowshipID": 1002, "FirstName": "Samwise", "Skills": "Gardening"},
    {"FellowshipID": 1003, "FirstName": "Gandalf", "Skills": "Spells"},
    {"FellowshipID": 1004, "FirstName": "Pippin",  "Skills": None}
]

# Create the DataFrame
df1 = pd.DataFrame(data)

# Display the DataFrame
print(df1)



# Define the data as a list of dictionaries
data2 = [
    {"FellowshipID": 1001, "FirstName": "Frodo",    "Age": 50},
    {"FellowshipID": 1002, "FirstName": "Samwise",  "Age": 39},
    {"FellowshipID": 1006, "FirstName": "Legolas",  "Age": 2931},
    {"FellowshipID": 1007, "FirstName": "Elrond",   "Age": 6520},
    {"FellowshipID": 1008, "FirstName": "Barromir", "Age": 51}
]

# Create the DataFrame
df2 = pd.DataFrame(data2)

# Display the DataFrame
print(df2)
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
# if we do not provide how and on columns also merge will compare the similar columns on the both dataframe and perform the join.
```

```python
df1.merge(df2, how = 'outer')
```

```python
# This method is primarily designed for combining DataFrames based on their indices. 
# By default, it attempts to join on the index of the calling DataFrame with the index of the other DataFrame.
# Also you need to provide suffix for left and right because it cannot understand which common columns in both the dataframe belongs to which dataframe.
join_df1 = df1.join(df2, lsuffix = '_Left', rsuffix = '_Right')
join_df1
```

```python
# if you see the output above, it joined based on index values on the both the dataframes but not on the column values. 
# so lets mention the join columns as well as below:
join_df2 = df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left',rsuffix = '_Right')
join_df2
# still the output doesnt look nice. because as mentioned above joins look nice on index rather than column values.
```

```python
# now lets do join on index
join_df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
join_df3
```

```python
# concat will simply combine pandas DataFrames, along a particular axis. This function is similar to the UNION ALL operation in SQL.
pd.concat([df1,df2]) #default axis = 0
```

```python
# as we mentioned axis =1 here, the data is combined based on columns rather than rows. 
# default join condition is outer
pd.concat([df1,df2], join = 'inner', axis = 1) 
```
